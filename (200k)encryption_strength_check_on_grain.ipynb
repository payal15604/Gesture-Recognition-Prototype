```python
import pandas as pd

# Step 1: Read the data from CSV
data = pd.read_csv('data2.csv')
```

```python
# Step 2: Convert iv and keystream columns from hexadecimal to binary
data['iv_binary'] = data['IV'].apply(lambda x: bin(int(x, 16))[2:].zfill(64))
data['keystream_binary'] = data['Keystream'].apply(lambda x: bin(int(x, 16))[2:].zfill(80))
```

```python
# Step 3: Get the structure of the DataFrame
print("DataFrame Info:")
print(data.info())
```

```python

# Display the first few rows of the DataFrame
print("\nFirst few rows of the DataFrame:")
print(data.head())
```

```python
# Assuming data is the DataFrame containing 'iv_binary' and 'keystream_binary' columns
all_samples_list = []
for index, row in data.iterrows():
    # Extract the first 64 elements (iv_binary) and the last element (first bit of keystream_binary)
    sample = list(row['iv_binary']) + [row['keystream_binary'][0]]
    all_samples_list.append(sample)
```

```python
# Assuming data is the DataFrame containing 'iv_binary' and 'keystream_binary' columns
all_samples_list = []
for index, row in data.iterrows():
    # Extract the first 64 elements (iv_binary) and the last element (first bit of keystream_binary)
    sample = list(row['iv_binary']) + [row['keystream_binary'][0]]
    all_samples_list.append(sample)
    print(sample)

print(all_samples_list)
```

```python
#split the all_samples_list into two separate lists: one containing the first 64
#bits (IV) and the other containing only the 65th bit (Keystream)
# Assuming all_samples_list contains the data with first 64 bits (IV) and the 65th bit (Keystream)
iv_list = [sample[:64] for sample in all_samples_list]
keystream_bit_list = [sample[64] for sample in all_samples_list]

# Print the first 10 elements of each list
print("IV List:")
print(iv_list[:10])
print("\nKeystream Bit List:")
print(keystream_bit_list[:10])

```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Step 4: Convert the list of list to a NumPy array
X = np.array(all_samples_list)
```

```python
# Step 5: Extract the first 64 bits of each sample (IV) as the input data
X_iv = X[:, :64].astype(float)  # Convert to float
```

```python
y = X[:, -1].astype(float)  # Convert to float
```

```python
# Step 7: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_iv, y, test_size=0.2, random_state=42)
```

```python
# Step 8: Build the neural network model
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(64,)))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# Step 9: Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Step 10: Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Step 11: Evaluate the model
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print("Test accuracy:", test_accuracy)
```

```python
# Step 12: Make predictions on new data (in this case, iv_list)
predictions = model.predict(X_iv)
print("Predictions:", predictions)
```
